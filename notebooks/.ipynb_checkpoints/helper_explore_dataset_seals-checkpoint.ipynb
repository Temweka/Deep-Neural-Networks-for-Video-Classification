# Visualize some of the data

```python
labels = pd.read_csv(path_data + 'labels.csv')
```

```python
labels[labels['split'] == 'valid']['video'].unique()
```

```python
labels[labels['split'] == 'test']['video'].unique()
```

```python
df = labels.groupby(['video','label']).agg(len).unstack(level=-1)
df.unstack(level=-1)
df.columns = df.columns.droplevel()
```

```python
df=df[['seal','noseal']]
df.columns = ["seal",'del','noseal','del']
del df['del']
```

```python
df.head()
```

```python
df['count'] = df['seal'] + df['noseal']
```

```python
df['seal_pct'] = df['seal'] / df['count']
```

```python
df.sort_values('video', inplace=True)
```

```python
# import matplotlib.pyplot as plt
df['seal_pct'].plot(kind='bar', figsize = (20,10), color = 'gray', title='Percentage of frames that contain a seal')
plt.show()
```

```python
df['seal_pct'].mean()
```

## view some frames

```python
from PIL import Image
```

```python
def view_video_sample(vid):
    paths = list(labels[labels['video'] == vid]['frame'])
    paths = [path_data + vid + '/' + p for p in paths]

    # plot first
    img = Image.open(paths[0])
    plt.imshow(img)
    plt.show()

    # plot last
    img = Image.open(paths[len(paths)-1])
    plt.imshow(img)
    plt.show()
```

```python
for vid in list(labels.sort_values('split')['video'].unique()):
    print(vid, list(labels[labels['video'] == vid].head(1)['split'])[0])
    view_video_sample(vid)
```
